```
import math

def merge(items, p, q, r):
    L = items[p:q+1]        #  Toma desde p hasta q
    R = items[q+1:r+1]      #  toma desde q+1 hasta r
    i = j = 0
    k = p
    while i < len(L) and j < len(R):
        if(L[i] < R[j]):
            items[k] = L[i]
            i += 1
        else:
            items[k] = R[j]
            j += 1
        k += 1
    if(j == len(R)):
        items[k:r+1] = L[i:]



def mergesort(items, p, r):
    if(p < r):
        q = math.floor((p+r)/2)
        print(items)
        mergesort(items, p, q)
        print(items)
        mergesort(items, q+1, r)
        print(items)
        merge(items, p, q, r)


items = [4,3,2,1,17]
mergesort(items, 0, len(items)-1)
items
```

```
def MERGE(A,p,q,r):
  n1 = q - p + 1    #   Elementos en A, desde p hasta q
  n2 = r - q        #   Elementos en A, desde q+1, hasta R
  L = np.zeros(n1+1) #  n1 + 1 elementos
  R = np.zeros(n2+1) #  n2 + 1 elementos
  for i in range(n1):
    L[i] = A[p + i]
  for j in range(n2):
    R[j] = A[q + j + 1]  
  i = 0
  j = 0
  for k in range(p,r+1):
    if L[i] <= R[j]:
      A[k] = L[i]
      i =+ 1
    else: 
      A[k] = R[j]
      j =+ 1
```
